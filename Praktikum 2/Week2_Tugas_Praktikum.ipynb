**4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.**

```python
# Import libraries yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load your dataframe (replace 'your_data.csv' with your actual file)
# Check if the file exists in the current directory. If not, provide the full path to the file.
df = pd.read_csv('/content/wbc.csv') # Assuming your data is in a CSV file

# Memilih kolom numerik (kecuali kolom non-numerik seperti 'id' dan 'diagnosis')
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Inisialisasi StandardScaler untuk standarisasi data numerik
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Memisahkan fitur (X) dan label target (y)
X = df.drop(columns=['id', 'diagnosis'])  # Fitur yang digunakan
y = df['diagnosis']  # Kolom target

# Melakukan stratified split dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Menampilkan hasil split
print("Jumlah data latih:", X_train.shape[0])
print("Jumlah data uji:", X_test.shape[0])
print("\nProporsi kelas di data latih:\n", y_train.value_counts(normalize=True))
print("\nProporsi kelas di data uji:\n", y_test.value_counts(normalize=True))
```
